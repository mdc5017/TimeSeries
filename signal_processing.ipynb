```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import scipy.signal
from scipy import *
import copy
%matplotlib inline
```

### Create artificial data with noise

```python
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (12,5)

sigRate = 1000 #Hz
time = np.arange(0,3,1/sigRate)
n = len(time)
p = 15 # poles for random interpolation
ampl = np.interp(np.linspace(0,p,n), np.arange(0,p), np.random.rand(p)*30)
plt.plot(time, ampl)
```

#### Add arbitrary noise

```python
noise_amp = 5
noise = noise_amp*np.random.randn(n)
signal = ampl + noise

plt.plot(time, signal, color='orange')
plt.plot(time, ampl, color='blue')
```

```python
# Initialising
filtSig = np.zeros(n)
k=20
for i in range(k, n-k-1):
    # each point is the average of k surrounding points
    filtSig[i] = np.mean(signal[i-k:i+k])

plt.plot(signal)
plt.plot(filtSig)
```
